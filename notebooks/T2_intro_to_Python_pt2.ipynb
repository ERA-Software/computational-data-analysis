**1.1. Library**: is a collection of non-volatile resources used by computer programs, often to manage data or perform complex calculations.
One useful example could be the library **math** that improve the readability of formula, moreover, it has preconfigured constants: π and e. It is part of python standard library, to exploit its features you must only move from disk to RAM by means `import`.

````python
# To use a library you have to "import" it from Disk memory to RAM
import math

# Let's calculate precisely a Area of a circle with R=1:
R = 1
Area = math.pi * R**2
print(f"The Area of the circle is: {Area}")

# Let's evaluate the square root of 16:
print(f"The square root of 16 is: {math.sqrt(16)}")
````

````python
# To save memory is also possible to import in the RAM only a module:
from math import sqrt

print(sqrt(25))         # 5.0
# print(pi)             # This would fail because we ONLY took sqrt!
````

````python
# Moreover, to save time you can use the short name:
import math as m

print(m.factorial(5))   # 120 (5*4*3*2*1)
````

## 2. **Functions**
A function is a black-box part of the code used to improve **readability** of the main/backbone code.
When python hits a function call, it doesn't just read the next line. It performs a Control Flow Jump: it "pauses" the main script, it "jumps" to the function block, it creates a Local Scope (a temporary memory space), runs the code and ,eventually, goes back to the exact spot left carrying the return value like a courier.
Depending on the complexity of the code, you can define the function before in the same source file and directly use it, or you can define it a different source file and **import** it.

**2.1. Syntax**
``` python
def myFunction(list_of_input):      # Order matters
---> """
     Block note usually exploited to explain the features of the
     function:
     - input:....
     - output:... Force [kN]..
     """
     myVariable = 1   # Valid only inside the function
     return list_of_output
```

````python
# A function to retrieve the max in between two numbers:

def myMax(a,b):
   if a>b:
     return a
   else:
     return b
a = 50
b = 5
print(myMax(a,b))
````

````python
# If I have to output more than one variables:
def toUpperLower(input_string):
    upper_string = input_string.upper()
    lower_string = input_string.lower()
    return upper_string, lower_string

myString = 'CiaoCIAOciao'
upper, lower = toUpperLower(myString)
print(upper)
print(lower)
uplow = toUpperLower(myString)
print(uplow)  # In this case I will obtain a tuple (Not changeble item)
````

````python
# Parameters of a function:
def myGreater(a,b=0):       # b=0 as default value, if nothing is specified
   return a>b
result = myGreater(-1)
print(result)

# It is possible to reassign the value of b:
result = myGreater(-1,-10)
print(result)
````

````python
# Named parameters:
def myFunction(a,b,c,d):
    print(a)
    print(b)
    print(c)
    print(d)
    return a+b+c+d

sum = myFunction(d=1,a=2,c=3,b=4)  # The order doesn't matter if you explicitly call the input variables
print(sum)

sum_2 = myFunction(1,2,3,4)        # In this case, values are assigned order-wise
print(sum_2)

d=4
sum_3 = myFunction(a=1, b=2, c=3, d=d)
print(sum_3)
````

**2.2. High-order function**
In the previous examples we gave variables as input to a function and retrive output variables. If the input and output of a functions are functions themself, this is called **high-order** computing. In the following, it is presented `map` and its syntax. This command will apply a function to all the elements of a list (It is actually better than a for loop because computation run in parallel). \
N.B. It is better than a for-loop beacuse it runs functions in paralell.

````python
# map(function, iterable,...) indicate the list of input to the functions
def toUpper(input_string):
    return input_string.upper()

myList = ['Halo','CIao']
upperList = map(toUpper,myList)
print(upperList)
print(list(upperList))
````

**2.3. Anonymous function**
It is called `lambda`: deep inside python threat it as a function and it returns a function object. Moreover, it is very useful for simple one-liner function (it avoids to define a function from scratch) and whose result is immediately returned. \
``` python
result = lambda parameters:expression
```


````python
# Convert a unit of measure:

diameter_cm = 20
diameter_inch = lambda x: x/2.54
print(f"The lambda fun. is useful to convert a measure from {diameter_cm} cm to {diameter_inch(diameter_cm)} inches")
````

````python
# Convert a list of measures:

diameter_cm = [20,30,40]
diameter_inch = list(map(lambda x: x/2.54, diameter_cm))   # Need to declare as a list()
print(diameter_inch)
````

````python
import kagglehub
path = kagglehub.dataset_download("chinmayshanbhag/big-basket-products")
````

````python
import os
import pandas as pd
# Database management DO NOT TOUCH THIS!
# List the contents of the downloaded dataset directory to find the CSV file
# The 'path' variable is available from the previous step
dataset_files = os.listdir(path)
csv_files = [f for f in dataset_files if f.endswith('.csv')]

if not csv_files:
    print(f"No CSV files found in the directory: {path}")
    csv_file_path = None
elif len(csv_files) == 1:
    csv_file_name = csv_files[0]
    csv_file_path = os.path.join(path, csv_file_name)
    print(f"Found CSV file: {csv_file_path}")
else:
    print(f"Multiple CSV files found: {csv_files}. Attempting to use the first one: {csv_files[0]}")
    csv_file_name = csv_files[0]
    csv_file_path = os.path.join(path, csv_file_name)

if csv_file_path:
    # Load the CSV file into a pandas DataFrame
    df = pd.read_csv(csv_file_path)

    print("\nFirst 5 rows of the DataFrame:")
    display(df.head(20))

    print("\nLast 5 rows of the DataFrame:")
    display(df.tail())
else:
    print("Could not proceed as no CSV file was found.")
````

````python
required_columns = ['ProductName', 'Price', 'DiscountPrice', 'Category']
df_filtered = df[required_columns].copy()

# Ensure prices are floats
df_filtered['Price'] = df_filtered['Price'].astype(float)

# Split 'Quantity' into 'qty' and 'unit_of_measure'
df_filtered[['qty', 'unit_of_measure']] = df['Quantity'].str.extract(r'(\d+\.?\d*)\s*([a-zA-Z]+)?')

# Convert 'qty' to numeric (float) and fill any NaN with 0
df_filtered['qty'] = pd.to_numeric(df_filtered['qty'], errors='coerce').fillna(0)

# Handle cases where unit might be missing (e.g., if only a number is present)
df_filtered['unit_of_measure'] = df_filtered['unit_of_measure'].fillna('')


display(df_filtered.head())
display(df_filtered.tail())
````

````python
# Convert the DataFrame to a list of dictionaries
df_as_dict = df_filtered.to_dict(orient='records')

# Display the first 5 entries of the dictionary list to verify
print("First 5 rows of df_filtered as a list of dictionaries:")
for i in range(min(5, len(df_as_dict))):
    print(df_as_dict[i])
````
