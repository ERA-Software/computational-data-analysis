## 2. **Functions**
A function is a black-box part of the code used to improve **readability** of the main/backbone code.
When python hits a functiob call, it doesn't just read the next line. It performs a Control Flow Jump: it "pauses" the main script, it "jumps" to the function block, it creates a Local Scope (a temporary memory space), runs the code and ,eventually, goes back to the exact spot left carrying the return value lika a courier.
Depending on the complexity of the code, you can define the function before in the same source file and directly use it, or you can define it ina different source file and **import** it.

````python
import math

def rebar_safety_check(diameter_cm, safety_coeff, force_N, E_MPa, yield_strength_MPa=350):
    """
    Performs a safety check for a rebar subjected to tensile forces.

    Args:
        diameter_cm (float): Diameter of the rebar in cm.
        safety_coeff (float): Safety coefficient.
        force_N (float): Applied tensile force in Newtons.
        E_MPa (float): Young's Modulus in MPa (N/mm^2).
        yield_strength_MPa (float): Yield strength of the material in MPa (N/mm^2).

    Returns:
        dict: A dictionary containing calculated stress, allowable stress, safety status, and strain.
    """
    # Convert diameter from cm to mm for consistency with MPa (N/mm^2)
    diameter_mm = diameter_cm * 10

    # Calculate cross-sectional area in mm^2
    area_mm2 = math.pi * (diameter_mm / 2)**2

    # Calculate actual stress in MPa
    if area_mm2 == 0:
        stress_MPa = float('inf') # Avoid division by zero
    else:
        stress_MPa = force_N / area_mm2

    # Calculate allowable stress in MPa
    allowable_stress_MPa = yield_strength_MPa / safety_coeff

    # Perform safety check
    is_safe = stress_MPa <= allowable_stress_MPa

    # Calculate strain (unitless)
    # Assuming stress is within elastic limit to use Hooke's Law
    strain = stress_MPa / E_MPa

    return {
        "diameter_cm": diameter_cm,
        "force_N": force_N,
        "stress_MPa": stress_MPa,
        "allowable_stress_MPa": allowable_stress_MPa,
        "is_safe": is_safe,
        "strain": strain
    }

print("EX.1: Rebar Safety Check")
# Input for EX.1
diameter_ex1 = 10          # [cm]
safety_coeff_ex1 = 1.15    # [-]
force_ex1 = 1000           # [N]
E_ex1 = 2e5                # [MPa]
yield_strength_ex1 = 350   # [MPa] - Assuming a typical steel yield strength

result_ex1 = rebar_safety_check(diameter_ex1, safety_coeff_ex1, force_ex1, E_ex1, yield_strength_ex1)

print(f"  Diameter: {result_ex1['diameter_cm']} cm")
print(f"  Applied Force: {result_ex1['force_N']} N")
print(f"  Calculated Stress: {result_ex1['stress_MPa']:.2f} MPa")
print(f"  Allowable Stress (with safety factor): {result_ex1['allowable_stress_MPa']:.2f} MPa")
print(f"  Strain: {result_ex1['strain']:.6f}")
print(f"  Is Safe: {result_ex1['is_safe']}")
````

**EX.3 Check with Gemini what is the "Inheritance" of a Class. Create a Class called 'Point2D' taking as input and attributes x and y coordinate of a point in Cartesian plane. Define a method called 'movePoint' that move P in the plane of dx and dy. Eventually use inheritance to call the class 'Point3D' that manages a point in 3D, updating as well the function movePoint.**

### **Solution to EX.3: Class Inheritance (Point2D and Point3D)**

````python
class Point2D:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def movePoint(self, dx, dy):
        self.x += dx
        self.y += dy

    def __str__(self):
        return f"Point2D({self.x}, {self.y})"

class Point3D(Point2D):
    def __init__(self, x, y, z):
        super().__init__(x, y)  # Initialize base class (Point2D) attributes
        self.z = z

    def movePoint(self, dx, dy, dz):
        super().movePoint(dx, dy) # Call base class movePoint for x and y
        self.z += dz

    def __str__(self):
        return f"Point3D({self.x}, {self.y}, {self.z})"

# Demonstration
print("--- Demonstrating Point2D ---")
p2d = Point2D(1, 2)
print(f"Initial 2D point: {p2d}")
p2d.movePoint(3, -1)
print(f"Moved 2D point by (3, -1): {p2d}")

print("\n--- Demonstrating Point3D (inherits from Point2D) ---")
p3d = Point3D(10, 20, 30)
print(f"Initial 3D point: {p3d}")
p3d.movePoint(-5, 5, 10)
print(f"Moved 3D point by (-5, 5, 10): {p3d}")
````

#### **Task 2**

Here, we use a `lambda` function for a concise way to compute the percentage discount for each product. We'll handle cases where the original price might be zero to avoid division by zero errors.

````python
# Use a lambda function to calculate percentage discount
percentage_discounts = []
for item in df_as_dict:
    price = item['Price']
    discount_price = item['DiscountPrice']
    discount_percentage = (lambda p, dp: ((p - dp) / p * 100) if p > 0 else 0)(price, discount_price)
    percentage_discounts.append({
        'ProductName': item['ProductName'],
        'OriginalPrice': price,
        'DiscountPrice': discount_price,
        'DiscountPercentage': f"{discount_percentage:.2f}%"
    })

print("First 5 items with percentage discounts:")
for i in range(min(5, len(percentage_discounts))):
    print(percentage_discounts[i])
````

````python
import math

# Part 1: Product Class
class Product:
    def __init__(self, product_name, category, price, discount_price, quantity=0.0, unit_of_measure='unit'):
        self.product_name = product_name
        self.category = category
        self.price = float(price)
        self.discount_price = float(discount_price)
        self.quantity = float(quantity)
        self.unit_of_measure = unit_of_measure

    def get_current_selling_price(self):
        if self.discount_price > 0 and self.discount_price < self.price:
            return self.discount_price
        return self.price

    def calculate_discount_percentage(self):
        if self.price > 0 and self.discount_price > 0 and self.discount_price < self.price:
            percentage = ((self.price - self.discount_price) / self.price) * 100
            return f"{percentage:.2f}%"
        return "0.00%"

    def update_stock(self, change):
        if self.quantity + change >= 0:
            self.quantity += change
            return True
        print(f"Error: Cannot reduce stock below zero for {self.product_name}. Current stock: {self.quantity} {self.unit_of_measure}")
        return False

    def display_info(self):
        print(f"\n--- Product: {self.product_name} ---")
        print(f"  Category: {self.category}")
        print(f"  Original Price: ${self.price:.2f}")
        print(f"  Discount Price: ${self.discount_price:.2f}")
        print(f"  Current Selling Price: ${self.get_current_selling_price():.2f}")
        print(f"  Discount Percentage: {self.calculate_discount_percentage()}")
        print(f"  Stock: {self.quantity:.2f} {self.unit_of_measure}")

# Part 2: GroceryInventoryManager Class
class GroceryInventoryManager:
    def __init__(self):
        self.inventory = []

    def add_product(self, product_obj):
        # Check if product already exists
        existing_product = self.find_product(product_obj.product_name)
        if existing_product:
            existing_product.update_stock(product_obj.quantity)
            print(f"Updated stock for {product_obj.product_name}. New quantity: {existing_product.quantity:.2f} {existing_product.unit_of_measure}")
        else:
            self.inventory.append(product_obj)
            print(f"Added new product: {product_obj.product_name} with quantity {product_obj.quantity:.2f} {product_obj.unit_of_measure}")

    def remove_product(self, product_name):
        original_len = len(self.inventory)
        self.inventory = [p for p in self.inventory if p.product_name != product_name]
        if len(self.inventory) < original_len:
            print(f"Removed product: {product_name}")
            return True
        print(f"Product '{product_name}' not found in inventory.")
        return False

    def find_product(self, product_name):
        for product in self.inventory:
            if product.product_name == product_name:
                return product
        return None

    def get_total_inventory_value(self):
        total_value = 0.0
        for product in self.inventory:
            total_value += product.get_current_selling_price() * product.quantity
        return total_value

    def list_products_by_category(self, category):
        return [p for p in self.inventory if p.category == category]

    def display_all_products(self):
        if not self.inventory:
            print("Inventory is currently empty.")
            return
        print("\n--- Current Grocery Inventory ---")
        for product in self.inventory:
            product.display_info()
        print("-----------------------------------")
````

````python
# Part 3: Implementation and Demonstration

# 1. Initialize Manager
grocery_manager = GroceryInventoryManager()

# 2. Populate Inventory from df_as_dict
print("\nPopulating inventory...")
for item_data in df_as_dict:
    product = Product(
        product_name=item_data['ProductName'],
        category=item_data['Category'],
        price=item_data['Price'],
        discount_price=item_data['DiscountPrice'],
        quantity=item_data['qty'],
        unit_of_measure=item_data['unit_of_measure']
    )
    grocery_manager.add_product(product)

# 3. Demonstrate Functionality

# Display initial state of the entire inventory
grocery_manager.display_all_products()

# Find and display information about a specific product
print("\n--- Demonstrating find_product and display_info ---")
onion_loose = grocery_manager.find_product('Onion (Loose)')
if onion_loose:
    onion_loose.display_info()
else:
    print("'Onion (Loose)' not found.")

# Simulate a sale: "sell" 1.5 kg of 'Onion (Loose)'
print("\n--- Simulating a sale (1.5 kg of 'Onion (Loose)') ---")
if onion_loose:
    if onion_loose.update_stock(-1.5):
        print("Sale successful.")
    onion_loose.display_info()

# Simulate a restock: "add" 10.0 kg of 'Onion (Loose)'
print("\n--- Simulating a restock (10.0 kg of 'Onion (Loose)') ---")
if onion_loose:
    if onion_loose.update_stock(10.0):
        print("Restock successful.")
    onion_loose.display_info()

# Attempt to sell more of a product than is available
print("\n--- Attempting to over-sell 'Onion (Loose)' (e.g., -100 kg) ---")
if onion_loose:
    onion_loose.update_stock(-100.0)
    onion_loose.display_info()

# Calculate and print the get_total_inventory_value() of the entire grocery store
print("\n--- Total Inventory Value ---")
print(f"Total value of current inventory: ${grocery_manager.get_total_inventory_value():.2f}")
````
