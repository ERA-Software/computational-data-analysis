## 2. **Functions**
A function is a black-box part of the code used to improve **readability** of the main/backbone code.
When python hits a functiob call, it doesn't just read the next line. It performs a Control Flow Jump: it "pauses" the main script, it "jumps" to the function block, it creates a Local Scope (a temporary memory space), runs the code and ,eventually, goes back to the exact spot left carrying the return value lika a courier.
Depending on the complexity of the code, you can define the function before in the same source file and directly use it, or you can define it ina different source file and **import** it.

````python
# If I have to output more than one variables:
def toUpperLower(input_string):
    upper_string = input_string.upper()
    lower_string = input_string.lower()
    return upper_string, lower_string

myString = 'CiaoCIAOciao'
upper, lower = toUpperLower(myString)
print(upper)
print(lower)
uplow = toUpperLower(myString)
print(uplow)  # In this case I will obtain a tuple (Not changeble item)
````

````python
# Parameters of a function:
def myGreater(a,b=0):       # b=0 as default value, if nothing is specified
   return a>b
result = myGreater(-1)
print(result)

# It is possible to reassign the value of b:
result = myGreater(-1,-10)
print(result)
````

````python
# Named parameters:
def myFunction(a,b,c,d):
    print(a)
    print(b)
    print(c)
    print(d)
    return a+b+c+d

sum = myFunction(d=1,a=2,c=3,b=4)  # The order doesn't matter if you explicitely call the input variables
print(sum)

sum_2 = myFunction(1,2,3,4)        # In this case the value are assigned order-wise
print(sum_2)

d=4
sum_3 = myFunction(a=1, b=2, c=3, d=d)
print(sum_3)
````

**2.2. High-order function**
In the previous examples we gave variables as input to a function and retrive output variables. If the input and output of a functions are functions themself, this is called **high-order** computing. In the following, it is presented `map` and its syntax. This command will apply a function to all the elements of a list (It is actually better than a for loop because computation run in parallel). \
N.B. It is better than a for-loop beacuse it runs functions in paralell.

````python
# map(function, iterable,...) indicate the list of input to the functions
def toUpper(input_string):
    return input_string.upper()

myList = ['Halo','CIao']
upperList = map(toUpper,myList)
print(upperList)
print(list(upperList))
````

#### **Task 1: Categorize items by price using a custom function**

We'll define a function that takes an item (dictionary) and returns a price category ('Budget', 'Standard', 'Premium') based on its `DiscountPrice`.

````python
def get_price_category(item):
    price = item['DiscountPrice']
    if price < 50:
        return 'Budget'
    elif 50 <= price < 150:
        return 'Standard'
    else:
        return 'Premium'

# Apply the custom function to each item
item_price_categories = []
for item in df_as_dict:
    item_price_categories.append({
        'ProductName': item['ProductName'],
        'Category': item['Category'],
        'DiscountPrice': item['DiscountPrice'],
        'PriceCategory': get_price_category(item)
    })

print("First 5 items with price categories:")
for i in range(min(5, len(item_price_categories))):
    print(item_price_categories[i])
````

#### **Task 2: Calculate the percentage discount for each item using a lambda function**

Here, we use a `lambda` function for a concise way to compute the percentage discount for each product. We'll handle cases where the original price might be zero to avoid division by zero errors.

````python
# Use a lambda function to calculate percentage discount
percentage_discounts = []
for item in df_as_dict:
    price = item['Price']
    discount_price = item['DiscountPrice']
    discount_percentage = (lambda p, dp: ((p - dp) / p * 100) if p > 0 else 0)(price, discount_price)
    percentage_discounts.append({
        'ProductName': item['ProductName'],
        'OriginalPrice': price,
        'DiscountPrice': discount_price,
        'DiscountPercentage': f"{discount_percentage:.2f}%"
    })

print("First 5 items with percentage discounts:")
for i in range(min(5, len(percentage_discounts))):
    print(percentage_discounts[i])
````

#### **Task 3: Create a list of formatted product descriptions using `map` with a lambda function**

This task demonstrates the `map` function, which applies a given function (in this case, a `lambda`) to all items in an iterable. We'll create a human-readable description for each product.

````python
# Use map with a lambda function to create formatted product descriptions
product_descriptions = list(map(
    lambda item: f"{item['ProductName']} ({item['qty']} {item['unit_of_measure']}) - {item['Category']}",
    df_as_dict
))

print("First 5 formatted product descriptions:")
for i in range(min(5, len(product_descriptions))):
    print(product_descriptions[i])
````

#### **Task 4: Build a `GroceryInventoryManager` System using Classes**

This is a challenging task that requires you to apply your knowledge of classes, attributes, methods, and potentially error handling to simulate a basic inventory management system for the grocery warehouse. You'll encapsulate product data and create an object to manage the entire inventory.

**Part 1: The `Product` Class**

Create a Python class named `Product` to represent individual items in the grocery store. Each `Product` object should have the following attributes and methods:

*   **Attributes:**
    *   `product_name` (string)
    *   `category` (string)
    *   `price` (float, original price)
    *   `discount_price` (float, discounted price)
    *   `quantity` (float, current stock level)
    *   `unit_of_measure` (string, e.g., 'kg', 'unit')

*   **Constructor (`__init__`):** Initialize all the above attributes. Ensure `quantity` defaults to `0` if not provided.

*   **Methods:**
    *   `get_current_selling_price(self)`: Returns the `discount_price` if it's less than the `price` and greater than 0; otherwise, returns the `price`. If `discount_price` is 0, it means no discount is applied.
    *   `calculate_discount_percentage(self)`: Calculates and returns the percentage discount based on `price` and `discount_price`. Return 0 if there's no discount or if `price` is 0 to avoid division by zero. Format as a string (e.g., "25.45%").
    *   `update_stock(self, change)`: Adds `change` to the current `quantity`. If `change` is negative, it simulates a sale. Ensure the `quantity` does not drop below zero. Return `True` if successful, `False` otherwise.
    *   `display_info(self)`: Prints a well-formatted string containing all the product's attributes, its current selling price, and its discount percentage.

**Part 2: The `GroceryInventoryManager` Class**

Create a Python class named `GroceryInventoryManager` that will manage a collection of `Product` objects.

*   **Attributes:**
    *   `inventory` (list, to store `Product` objects)

*   **Constructor (`__init__`):** Initialize `inventory` as an empty list.

*   **Methods:**
    *   `add_product(self, product_obj)`: Adds a `Product` object to the `inventory` list. Before adding, check if a product with the same `product_name` already exists. If it does, update its quantity (add to existing stock) instead of adding a duplicate.
    *   `remove_product(self, product_name)`: Removes a `Product` object from the `inventory` by its `product_name`. Handle cases where the product is not found.
    *   `find_product(self, product_name)`: Searches the `inventory` for a product by its `product_name` and returns the `Product` object if found, otherwise returns `None`.
    *   `get_total_inventory_value(self)`: Calculates the sum of (current selling price * quantity) for all products in the inventory. Returns the total value as a float.
    *   `list_products_by_category(self, category)`: Returns a list of `Product` objects that belong to the specified `category`.
    *   `display_all_products(self)`: Iterates through the `inventory` and calls `display_info()` for each `Product` object.

**Part 3: Implementation and Demonstration**

1.  **Initialize Manager:** Create an instance of `GroceryInventoryManager`.
2.  **Populate Inventory:** Use the `df_as_dict` data (from previous Hackathon tasks) to create `Product` objects and add them to your `GroceryInventoryManager` using the `add_product` method. Make sure to handle the `qty` and `unit_of_measure` from `df_as_dict` to initialize the `quantity` and `unit_of_measure` attributes of your `Product` objects.
3.  **Demonstrate Functionality:**
    *   Display the initial state of the entire inventory.
    *   Find and display information about a specific product (e.g., 'Onion (Loose)').
    *   Simulate a sale: "sell" 1.5 kg of 'Onion (Loose)' and then display its updated info.
    *   Simulate a restock: "add" 10.0 kg of 'Onion (Loose)' and display its updated info.
    *   Attempt to sell more of a product than is available and observe the behavior.
    *   Calculate and print the `get_total_inventory_value()` of the entire grocery store.
  

````python
import math

# Part 1: Product Class
class Product:
    def __init__(self, product_name, category, price, discount_price, quantity=0.0, unit_of_measure='unit'):
        self.product_name = product_name
        self.category = category
        self.price = float(price)
        self.discount_price = float(discount_price)
        self.quantity = float(quantity)
        self.unit_of_measure = unit_of_measure

    def get_current_selling_price(self):
        if self.discount_price > 0 and self.discount_price < self.price:
            return self.discount_price
        return self.price

    def calculate_discount_percentage(self):
        if self.price > 0 and self.discount_price > 0 and self.discount_price < self.price:
            percentage = ((self.price - self.discount_price) / self.price) * 100
            return f"{percentage:.2f}%"
        return "0.00%"

    def update_stock(self, change):
        if self.quantity + change >= 0:
            self.quantity += change
            return True
        print(f"Error: Cannot reduce stock below zero for {self.product_name}. Current stock: {self.quantity} {self.unit_of_measure}")
        return False

    def display_info(self):
        print(f"\n--- Product: {self.product_name} ---")
        print(f"  Category: {self.category}")
        print(f"  Original Price: ${self.price:.2f}")
        print(f"  Discount Price: ${self.discount_price:.2f}")
        print(f"  Current Selling Price: ${self.get_current_selling_price():.2f}")
        print(f"  Discount Percentage: {self.calculate_discount_percentage()}")
        print(f"  Stock: {self.quantity:.2f} {self.unit_of_measure}")

# Part 2: GroceryInventoryManager Class
class GroceryInventoryManager:
    def __init__(self):
        self.inventory = []

    def add_product(self, product_obj):
        # Check if product already exists
        existing_product = self.find_product(product_obj.product_name)
        if existing_product:
            existing_product.update_stock(product_obj.quantity)
            print(f"Updated stock for {product_obj.product_name}. New quantity: {existing_product.quantity:.2f} {existing_product.unit_of_measure}")
        else:
            self.inventory.append(product_obj)
            print(f"Added new product: {product_obj.product_name} with quantity {product_obj.quantity:.2f} {product_obj.unit_of_measure}")

    def remove_product(self, product_name):
        original_len = len(self.inventory)
        self.inventory = [p for p in self.inventory if p.product_name != product_name]
        if len(self.inventory) < original_len:
            print(f"Removed product: {product_name}")
            return True
        print(f"Product '{product_name}' not found in inventory.")
        return False

    def find_product(self, product_name):
        for product in self.inventory:
            if product.product_name == product_name:
                return product
        return None

    def get_total_inventory_value(self):
        total_value = 0.0
        for product in self.inventory:
            total_value += product.get_current_selling_price() * product.quantity
        return total_value

    def list_products_by_category(self, category):
        return [p for p in self.inventory if p.category == category]

    def display_all_products(self):
        if not self.inventory:
            print("Inventory is currently empty.")
            return
        print("\n--- Current Grocery Inventory ---")
        for product in self.inventory:
            product.display_info()
        print("-----------------------------------")
````

````python
# Part 3: Implementation and Demonstration

# 1. Initialize Manager
grocery_manager = GroceryInventoryManager()

# 2. Populate Inventory from df_as_dict
print("\nPopulating inventory...")
for item_data in df_as_dict:
    product = Product(
        product_name=item_data['ProductName'],
        category=item_data['Category'],
        price=item_data['Price'],
        discount_price=item_data['DiscountPrice'],
        quantity=item_data['qty'],
        unit_of_measure=item_data['unit_of_measure']
    )
    grocery_manager.add_product(product)

# 3. Demonstrate Functionality

# Display initial state of the entire inventory
grocery_manager.display_all_products()

# Find and display information about a specific product
print("\n--- Demonstrating find_product and display_info ---")
onion_loose = grocery_manager.find_product('Onion (Loose)')
if onion_loose:
    onion_loose.display_info()
else:
    print("'Onion (Loose)' not found.")

# Simulate a sale: "sell" 1.5 kg of 'Onion (Loose)'
print("\n--- Simulating a sale (1.5 kg of 'Onion (Loose)') ---")
if onion_loose:
    if onion_loose.update_stock(-1.5):
        print("Sale successful.")
    onion_loose.display_info()

# Simulate a restock: "add" 10.0 kg of 'Onion (Loose)'
print("\n--- Simulating a restock (10.0 kg of 'Onion (Loose)') ---")
if onion_loose:
    if onion_loose.update_stock(10.0):
        print("Restock successful.")
    onion_loose.display_info()

# Attempt to sell more of a product than is available
print("\n--- Attempting to over-sell 'Onion (Loose)' (e.g., -100 kg) ---")
if onion_loose:
    onion_loose.update_stock(-100.0)
    onion_loose.display_info()

# Calculate and print the get_total_inventory_value() of the entire grocery store
print("\n--- Total Inventory Value ---")
print(f"Total value of current inventory: ${grocery_manager.get_total_inventory_value():.2f}")
````
