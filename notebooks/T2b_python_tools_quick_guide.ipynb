## Part 1 — Installing Packages: `!pip` vs `%pip`

When working in a Jupyter notebook (or Google Colab), you sometimes need to install a library before you can import it.
There are two ways to do this, and they are **not the same**.

### `!pip` — runs a shell command
The `!` prefix sends the command to your **operating system**, like opening a terminal and typing `pip install ...`.  
The problem: it might install the package into a *different* Python environment than the one your notebook is using.

### `%pip` — runs a notebook magic command
The `%` prefix is a built-in Jupyter command that installs the package directly into the **same Python kernel your notebook is running**.  
This is the safer, recommended choice inside notebooks.

```
Rule of thumb:
  Inside a notebook  →  use %pip install <package>
  Inside a terminal  →  activate your environment first, then use pip install <package>
```

````python
# Recommended: install a package into the current notebook kernel
# %pip install emoji

# Also works, but may target the wrong Python environment:
# !pip install emoji

# After installation, import as usual:
# import emoji
print("Use %pip inside notebooks — it always targets the right Python kernel.")
````

## Part 2 — OOP vs Functional Programming

There are two popular styles for organising Python code.

### Object-Oriented Programming (OOP)
Group **data** and **behaviour** together inside a `class`.  
Think of a class as a blueprint: it describes what an object *is* (its attributes) and what it can *do* (its methods).

**Good for:** modelling real-world things (a sensor, a building, a student), reusable components, large codebases.

### Functional Programming (FP)
Write **pure functions** that take inputs and return outputs — no shared state, no side effects.  
Data flows through a chain of functions.

**Good for:** data transformations, pipelines, short scripts, when you want predictable, testable logic.

> Most Python programs mix both styles. OOP for structure, FP for data processing.

````python
# ── OOP example ───────────────────────────────────────────────────────────────
class Sensor:
    """A temperature sensor with a location and a list of readings."""

    def __init__(self, location):
        self.location = location   # attribute: where the sensor is
        self.readings = []         # attribute: collected measurements

    def record(self, value):       # method: add a reading
        self.readings.append(value)

    def average(self):             # method: compute mean
        return sum(self.readings) / len(self.readings)

# Create two sensor objects from the same blueprint
s1 = Sensor("Bridge North")
s1.record(22.1)
s1.record(23.5)
s1.record(21.8)

print(f"Sensor: {s1.location}")
print(f"Readings: {s1.readings}")
print(f"Average: {s1.average():.1f} °C")
````

````python
# ── Functional Programming example ────────────────────────────────────────────
# Same problem — but now as plain functions, no class needed.

def average(readings):
    """Return the mean of a list of numbers."""
    return sum(readings) / len(readings)

def above_threshold(readings, threshold):
    """Return only the readings that exceed the threshold."""
    return [r for r in readings if r > threshold]

readings = [22.1, 23.5, 21.8, 25.0, 19.9]

print(f"Average: {average(readings):.1f} °C")
print(f"Readings above 22 °C: {above_threshold(readings, 22)}")

# Data flows through a pipeline of functions — easy to test each step in isolation
````

**Side-by-side comparison**

| | OOP | Functional |
|---|---|---|
| Core idea | Bundle data + behaviour in a `class` | Transform data through pure functions |
| State | Stored inside objects (`self.x`) | Passed as arguments, returned as values |
| When to use | Modelling entities, reusable components | Data pipelines, transformations |
| Example | `sensor.record(22.1)` | `average(readings)` |

## Part 3 — `lambda` and `map`

### `lambda` — a function with no name

A `lambda` is a shorthand for a simple, one-line function.  
Use it when you need a quick throwaway function and a full `def` would be overkill.

```python
# Regular function
def square(x):
    return x ** 2

# Equivalent lambda
square = lambda x: x ** 2
```

**Syntax:** `lambda <arguments>: <expression>`  
The expression is automatically returned — no `return` keyword needed.

### `map` — apply a function to every item in a list

`map(function, iterable)` applies `function` to each element and returns a map object.  
Wrap it in `list()` to get a plain list back.

````python
# ── lambda basics ─────────────────────────────────────────────────────────────
double = lambda x: x * 2
print(double(5))           # 10

add = lambda x, y: x + y
print(add(3, 4))           # 7

# Lambda inline — no variable needed
print((lambda x: x ** 2)(6))   # 36
````

````python
# ── map basics ────────────────────────────────────────────────────────────────
temps_celsius = [0, 20, 37, 100]

# Convert each temperature to Fahrenheit using map + lambda
temps_fahrenheit = list(map(lambda c: c * 9/5 + 32, temps_celsius))
print(temps_fahrenheit)   # [32.0, 68.0, 98.6, 212.0]

# map with a regular named function works too
def celsius_to_kelvin(c):
    return c + 273.15

temps_kelvin = list(map(celsius_to_kelvin, temps_celsius))
print(temps_kelvin)       # [273.15, 293.15, 310.15, 373.15]
````

````python
# ── lambda + map vs list comprehension ────────────────────────────────────────
# Both do the same thing — list comprehensions are often more readable in Python

values = [1, 2, 3, 4, 5]

# Using map + lambda
squared_map = list(map(lambda x: x ** 2, values))

# Using list comprehension (more Pythonic for simple cases)
squared_comp = [x ** 2 for x in values]

print(squared_map)   # [1, 4, 9, 16, 25]
print(squared_comp)  # [1, 4, 9, 16, 25]

# When is lambda+map preferred?
# When passing a function as an argument to another function (e.g., sorted, filter)
names = ["Alice", "charlie", "Bob"]
sorted_names = sorted(names, key=lambda name: name.lower())
print(sorted_names)  # ['Alice', 'Bob', 'charlie']
````

## Part 4 — Reading Files with Pandas

Pandas is the go-to library for loading and exploring tabular data.
It reads CSV and JSON files into a **DataFrame** — think of it as a spreadsheet inside Python.

### Key methods at a glance

| Method | What it does |
|--------|-------------|
| `pd.read_csv('file.csv')` | Load a CSV file into a DataFrame |
| `pd.read_json('file.json')` | Load a JSON file into a DataFrame |
| `df.head(n)` | Show first `n` rows (default 5) |
| `df.tail(n)` | Show last `n` rows (default 5) |
| `df.iloc[row, col]` | Select by **integer position** (row and column numbers) |
| `df.shape` | Number of (rows, columns) |
| `df.columns` | List of column names |

````python
import pandas as pd
import io

# ── We create a small CSV in memory so this notebook is self-contained ─────────
csv_data = """station,date,temp_C,rain_mm
A,2024-01-01,12.3,0.0
A,2024-01-02,11.8,2.5
A,2024-01-03,10.1,8.0
B,2024-01-01,15.6,0.0
B,2024-01-02,14.9,1.2
B,2024-01-03,13.7,5.5
"""

# Normally you would write: df = pd.read_csv('your_file.csv')
df = pd.read_csv(io.StringIO(csv_data))

print("Shape:", df.shape)        # (rows, columns)
print("Columns:", list(df.columns))
````

````python
# ── head and tail ──────────────────────────────────────────────────────────────
print("First 3 rows:")
print(df.head(3))

print("\nLast 2 rows:")
print(df.tail(2))
````

````python
# ── iloc — select by position ──────────────────────────────────────────────────
# df.iloc[row_index, column_index]  — both are zero-based integers

print(df.iloc[0])          # entire first row (as a Series)
print()
print(df.iloc[0, 2])       # row 0, column 2  → temp_C of first record
print()
print(df.iloc[0:3, 0:3])   # rows 0–2, columns 0–2 (slicing works like lists)
````

````python
# ── Reading JSON ───────────────────────────────────────────────────────────────
import json

# JSON is commonly used for API responses and configuration files.
# Pandas expects a list of records (each record is a dict).
json_data = json.dumps([
    {"station": "A", "date": "2024-01-01", "temp_C": 12.3},
    {"station": "A", "date": "2024-01-02", "temp_C": 11.8},
    {"station": "B", "date": "2024-01-01", "temp_C": 15.6},
])

# Normally: df_json = pd.read_json('your_file.json')
df_json = pd.read_json(io.StringIO(json_data))
print(df_json)
````
