## 5. **HACKATON: the grocery warehouse** 📦
A little local grocery store hires you as accountant. They provide to you a list of items in the warehouse and you are in charge of retrievieing data from the register and assist the owner to restock, calculate tax and other stuffs...
Luckly, you attended the Computational Data Analysis with Python course and you solve all the tasks of the week in 15 Minutes!!

```python
# This part of the code is genereting the ficticious dataset, DO NOT TOUCH IT!!
# THE DATASET: Copy and run this once.
warehouse = {}
categories = ["Food", "Beverage", "Fitness", "Tech", "Cleaning"]

for i in range(1, 101):
    item_id = f"SKU_{i:03d}"

    # Deterministic values based on the index 'i'
    # This ensures everyone has the same price and stock
    price = round((i * 1.25) % 50 + 0.99, 2)
    stock = (i * 7) % 101
    category = categories[i % 5]

    warehouse[item_id] = {
        "name": f"Product_{i}",
        "price": price,
        "stock": stock,
        "category": category
    }

print(f"✅ Warehouse initialized with {len(warehouse)} items.")
print(f"Example SKU_050: {warehouse['SKU_050']}")
```

Your Boss asks you:
1. To retrieve all the most expensive products in the category "Tech", with a price higher than 25 Euro.
2. The first product in the dictionary with a price lower than 10 Euros.
3. A costumer has 100 Euros, How many SKU_022 can buy?
4. To evaluate the total cost of the entire warehouse, the average price of a beverage and the cheapest food item.
5. Next week is a discount week, add a key to the dictionary called "Discount_price". Consider 30% discount for "Food", 20% for "Beverage" and 10% for all the other category.

### Task 1: Most Expensive 'Tech' Products (Price > 25 Euro)

```python
# 1. Retrieve all the most expensive products in the category 'Tech', with a price higher than 25 Euro.
expensive_tech_products = {}

for sku, details in warehouse.items():
    if details['category'] == 'Tech' and details['price'] > 25:
        expensive_tech_products[sku] = details

print("Most Expensive Tech Products (Price > 25 Euro):")
for sku, details in expensive_tech_products.items():
    print(f"  {sku}: {details['name']} - Price: {details['price']} Euro")
```

### Task 2: First Product with Price Lower Than 10 Euro

```python
# 2. The first product in the dictionary with a price lower than 10 Euros.
first_cheap_product = None
for sku, details in warehouse.items():
    if details['price'] < 10:
        first_cheap_product = (sku, details)
        break # Stop after finding the first one

if first_cheap_product:
    sku, details = first_cheap_product
    print(f"First product with price lower than 10 Euro: {sku} ({details['name']}) - Price: {details['price']} Euro")
else:
    print("No product found with a price lower than 10 Euro.")
```

### Task 3: How many SKU_022 can a customer buy with 100 Euros?

```python
# 3. A customer has 100 Euros, How many SKU_022 can buy?
customer_money = 100
sku_to_check = "SKU_022"

if sku_to_check in warehouse:
    item_price = warehouse[sku_to_check]['price']
    quantity_can_buy = int(customer_money / item_price)
    print(f"With {customer_money} Euro, a customer can buy {quantity_can_buy} units of {sku_to_check} (price: {item_price} Euro each).")
else:
    print(f"Product {sku_to_check} not found in the warehouse.")
```

### Task 4: Total Warehouse Cost, Average Beverage Price, Least Expensive Food Item

```python
# 4. To evaluate the total cost of the entire warehouse, the average price of a beverage and the less expensive food item.

# Calculate total cost of the entire warehouse
total_warehouse_cost = 0
for sku, details in warehouse.items():
    total_warehouse_cost += details['price'] * details['stock']
print(f"Total cost of the entire warehouse: {total_warehouse_cost:.2f} Euro")

# Calculate average price of a beverage
beverage_prices = []
for sku, details in warehouse.items():
    if details['category'] == 'Beverage':
        beverage_prices.append(details['price'])

if beverage_prices:
    average_beverage_price = sum(beverage_prices) / len(beverage_prices)
    print(f"Average price of a beverage: {average_beverage_price:.2f} Euro")
else:
    print("No beverage items found.")

# Find the least expensive food item
least_expensive_food_item = None
min_food_price = float('inf') # Initialize with a very high number

for sku, details in warehouse.items():
    if details['category'] == 'Food':
        if details['price'] < min_food_price:
            min_food_price = details['price']
            least_expensive_food_item = (sku, details)

if least_expensive_food_item:
    sku, details = least_expensive_food_item
    print(f"Least expensive food item: {sku} ({details['name']}) - Price: {details['price']} Euro")
else:
    print("No food items found.")
```

### Task 5: Add 'Discount_price' with category-specific discounts

```python
# 5. Next week is a discount week, add a key to the dictionary called "Discount_price".
# Consider 30% discount for "Food", 20% for "Beverage" and 10% for all the other category.

for sku, details in warehouse.items():
    original_price = details['price']
    category = details['category']

    if category == 'Food':
        discount_factor = 0.70  # 30% discount
    elif category == 'Beverage':
        discount_factor = 0.80  # 20% discount
    else:
        discount_factor = 0.90  # 10% discount for others

    details['Discount_price'] = round(original_price * discount_factor, 2)

print("Warehouse updated with 'Discount_price' for each item (showing first 5 items as example):")
# Print first 5 items to show the new key
for i, (sku, details) in enumerate(warehouse.items()):
    if i >= 5:
        break
    print(f"  {sku}: Price: {details['price']} Euro, Discount Price: {details['Discount_price']} Euro, Category: {details['category']}")

# Verify a specific item (e.g., SKU_050) to check the discount calculation
print(f"\nExample SKU_050 details with discount: {warehouse['SKU_050']}")
```
