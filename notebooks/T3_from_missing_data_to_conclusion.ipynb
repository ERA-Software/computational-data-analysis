## Why Time Series Data Processing Matters in Engineering

Modern engineering infrastructure — bridges, buildings, roads, water networks — is continuously monitored by sensors that record measurements over time: strain, deflection, temperature, traffic counts, water levels, and more. This produces **time series data**: sequences of values each stamped with a precise timestamp.

Being able to process this data is a core engineering skill because:

- **Sensors generate far more data than humans can inspect manually.** A single bridge sensor sampling once per second produces 86 400 readings per day. Computation is the only practical way to find patterns, anomalies, and trends.
- **Real data is messy.** Sensors go offline, transmissions fail, and readings get corrupted. Engineers must detect and handle these gaps before drawing any conclusions.
- **Engineering decisions depend on temporal context.** A deflection reading of 3 mm means something very different at 6 am (light traffic) versus 6 pm (rush hour) or in January (cold, steel contracted) versus July (hot, steel expanded). Without timestamps, the numbers are incomplete.
- **Trends and seasonality drive design and maintenance.** Identifying peak load periods, long-term deterioration, or anomalous events requires aggregating and smoothing data across time — not just reading individual values.

### Generate the Dataset

Run the cell below to create your dataset.
You do not need to understand every line — just run it and move on.

````python
np.random.seed(2024)
n_hours     = 8760   # 365 days × 24 hours
dates_2023  = pd.date_range(start='2023-01-01', periods=n_hours, freq='h')

# --- Traffic pattern ---
hour_of_day  = np.array([d.hour      for d in dates_2023])
day_of_week  = np.array([d.dayofweek for d in dates_2023])   # 0=Mon, 6=Sun

morning_peak   = 500 * np.exp(-0.5 * ((hour_of_day - 8)  / 1.5) ** 2)
evening_peak   = 450 * np.exp(-0.5 * ((hour_of_day - 17) / 1.5) ** 2)
weekend_factor = np.where(day_of_week >= 5, 0.55, 1.0)

traffic_signal = (80 + morning_peak + evening_peak) * weekend_factor
traffic_raw    = np.maximum(10, traffic_signal + np.random.normal(0, 25, n_hours)).astype(float)

# --- Temperature pattern ---
day_of_year = np.array([d.dayofyear for d in dates_2023])
seasonal    = 12 * np.sin(2 * np.pi * (day_of_year - 80) / 365)   # warm summer, cold winter
diurnal     =  6 * np.sin(2 * np.pi * (hour_of_day  - 6) / 24)    # warm afternoon, cool night
temp_raw    = np.round(13 + seasonal + diurnal + np.random.normal(0, 1.5, n_hours), 1)

# --- Introduce missing values (~3%) ---
for arr in [traffic_raw, temp_raw]:
    idx = np.random.choice(n_hours, size=int(0.03 * n_hours), replace=False)
    arr[idx] = np.nan

# --- Create Series ---
traffic_2023 = pd.Series(traffic_raw, index=dates_2023, name='vehicles_per_hour')
temp_2023    = pd.Series(temp_raw,    index=dates_2023, name='deck_temp_C')

print("Dataset ready!")
print(f"  Traffic:     {len(traffic_2023):,} hourly records  |  Missing: {traffic_2023.isnull().sum()}")
print(f"  Temperature: {len(temp_2023):,} hourly records  |  Missing: {temp_2023.isnull().sum()}")
````

````python
# ✅ SOLUTION — Task 2(e) ARIMA Imputation (Hard Extension)

from statsmodels.tsa.arima.model import ARIMA
import warnings

# Step 1: Resample to daily averages (ARIMA on 8760 hourly points is too slow)
traffic_daily_raw = traffic_2023.resample('D').mean()

# Step 2: Pre-fill so ARIMA has a complete series to fit on
traffic_prefilled = traffic_daily_raw.ffill().bfill()

# Step 3: Fit ARIMA(2,1,2) — captures trend (d=1) + short-range autocorrelation
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    model  = ARIMA(traffic_prefilled, order=(2, 1, 2))
    result = model.fit()

# Step 4: Use in-sample fitted values to fill only the missing slots
fitted = result.fittedvalues
traffic_arima_filled = traffic_daily_raw.copy()
traffic_arima_filled[traffic_daily_raw.isna()] = fitted[traffic_daily_raw.isna()]

# Step 5: Compare ARIMA vs. linear interpolation
traffic_interp = traffic_daily_raw.interpolate(method='linear')

ax = traffic_daily_raw.plot(
    alpha=0.5, label='Original (with gaps)', color='gray',
    title='Missing Value Imputation — Daily Traffic',
    ylabel='Vehicles / hour (daily avg)', figsize=(12, 4)
)
traffic_interp.plot(ax=ax, label='Linear interpolation', color='steelblue', linewidth=1.5)
traffic_arima_filled.plot(ax=ax, label='ARIMA imputation', color='red', linewidth=1.5, linestyle='--')
ax.legend()

print(f"ARIMA AIC: {result.aic:.1f}  |  Missing slots filled: {traffic_daily_raw.isna().sum()}")
print("Note: ARIMA and linear interpolation are often close for short gaps.")
````
