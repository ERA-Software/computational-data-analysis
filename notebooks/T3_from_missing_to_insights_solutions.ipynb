## Why Data Processing Matters in Engineering

Modern engineering infrastructure — bridges, buildings, roads, water networks — is continuously monitored by sensors that record measurements over time: strain, deflection, temperature, traffic counts, water levels, and more. This produces **time series data**: sequences of values each stamped with a precise timestamp.

Being able to process this data is a core engineering skill because:

- **Sensors generate far more data than humans can inspect manually.** A single bridge sensor sampling once per second produces 86 400 readings per day. Computation is the only practical way to find patterns, anomalies, and trends.
- **Real data is messy.** Sensors go offline, transmissions fail, readings get corrupted, and sometimes the same record is logged twice. Engineers must detect and handle these problems before drawing any conclusions.
- **Engineering decisions depend on temporal context.** A deflection reading of 3 mm means something very different at 6 am (light traffic) versus 6 pm (rush hour) or in January (cold, steel contracted) versus July (hot, steel expanded). Without timestamps, the numbers are incomplete.
- **Outliers and duplicates distort statistics.** A single corrupted reading of 99 999 vehicles/hour can shift an annual average. Duplicate records inflate totals. Identifying and handling these is essential before any analysis.

### Generate the Dataset

Run the cell below to create your dataset.
You do not need to understand every line — just run it and move on.

> The dataset for the following tasks is named `bridge_df`.

````python
np.random.seed(2024)
n_hours     = 8760   # 365 days × 24 hours
dates_2023  = pd.date_range(start='2023-01-01', periods=n_hours, freq='h')

# --- Traffic pattern ---
hour_of_day  = np.array([d.hour      for d in dates_2023])
day_of_week  = np.array([d.dayofweek for d in dates_2023])   # 0=Mon, 6=Sun

morning_peak   = 500 * np.exp(-0.5 * ((hour_of_day - 8)  / 1.5) ** 2)
evening_peak   = 450 * np.exp(-0.5 * ((hour_of_day - 17) / 1.5) ** 2)
weekend_factor = np.where(day_of_week >= 5, 0.55, 1.0)

traffic_signal = (80 + morning_peak + evening_peak) * weekend_factor
traffic_raw    = np.maximum(10, traffic_signal + np.random.normal(0, 25, n_hours)).astype(float)

# --- Temperature pattern ---
day_of_year = np.array([d.dayofyear for d in dates_2023])
seasonal    = 12 * np.sin(2 * np.pi * (day_of_year - 80) / 365)   # warm summer, cold winter
diurnal     =  6 * np.sin(2 * np.pi * (hour_of_day  - 6) / 24)    # warm afternoon, cool night
temp_raw    = np.round(13 + seasonal + diurnal + np.random.normal(0, 1.5, n_hours), 1)

# --- Introduce missing values (~3%) ---
for arr in [traffic_raw, temp_raw]:
    idx = np.random.choice(n_hours, size=int(0.03 * n_hours), replace=False)
    arr[idx] = np.nan

# --- Introduce outliers (~0.5%) ---
# Traffic outliers: implausibly high readings (sensor glitch sending max value)
outlier_idx_traffic = np.random.choice(n_hours, size=int(0.005 * n_hours), replace=False)
traffic_raw[outlier_idx_traffic] = np.random.uniform(2000, 5000, len(outlier_idx_traffic))

# Temperature outliers: extreme spikes (e.g. sensor malfunction)
outlier_idx_temp = np.random.choice(n_hours, size=int(0.005 * n_hours), replace=False)
temp_raw[outlier_idx_temp] = np.random.choice([-40, -35, 55, 60, 65], len(outlier_idx_temp))

# --- Build DataFrame ---
bridge_df = pd.DataFrame({
    'timestamp': dates_2023,
    'traffic': traffic_raw,
    'temperature': temp_raw
})

# --- Introduce duplicate rows (~1%) ---
dup_idx = np.random.choice(n_hours, size=int(0.01 * n_hours), replace=False)
duplicates = bridge_df.iloc[dup_idx].copy()
bridge_df = pd.concat([bridge_df, duplicates], ignore_index=True)
bridge_df = bridge_df.sort_values('timestamp').reset_index(drop=True)

print("Dataset ready!")
````

````python
# ✅ SOLUTION — Task 3

# (a) Detect and remove duplicates
# Setting a DatetimeIndex and aggregating by hour may create spurious duplicates
# due to coincidental matches in floating-point values.

# To prevent this, remove duplicates before setting the DatetimeIndex and performing hourly grouping:

# n_dups = bridge_df.duplicated().sum()
# print(f"Duplicate rows found: {n_dups}")
# print("\nExample duplicates:")
# print(bridge_df[bridge_df.duplicated(keep=False)].head(6))

# bridge_clean = bridge_df.drop_duplicates()
# print(f"\nAfter deduplication: {len(bridge_clean)} rows (removed {n_dups})")

# Alternative approach:
# If duplicates are introduced after setting the DatetimeIndex or during hourly grouping,
# they can be resolved using the following method:

n_dups = bridge_df.index.duplicated().sum()
bridge_clean = bridge_df[~bridge_df.index.duplicated(keep='first')]

print(f"\nAfter deduplication: {len(bridge_clean)} rows (removed {n_dups})")
print("\n--- Summary statistics ---")
print(bridge_clean[['traffic', 'temperature']].describe().round(1))

# (b) Detect outliers using IQR method
for col in ['traffic', 'temperature']:
    Q1 = bridge_clean[col].quantile(0.25)
    Q3 = bridge_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = bridge_clean[(bridge_clean[col] < lower) | (bridge_clean[col] > upper)]
    print(f"\n{col}: Q1={Q1:.1f}, Q3={Q3:.1f}, IQR={IQR:.1f}")
    print(f"  Bounds: [{lower:.1f}, {upper:.1f}]")
    print(f"  Outliers: {len(outliers)} ({len(outliers)/len(bridge_clean)*100:.2f}%)")

# (c) Handle outliers — replace with NaN (so they get filled in Task 4)
for col in ['traffic', 'temperature']:
    Q1 = bridge_clean[col].quantile(0.25)
    Q3 = bridge_clean[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    mask = (bridge_clean[col] < lower) | (bridge_clean[col] > upper)
    bridge_clean.loc[mask, col] = np.nan

print(f"\nAfter replacing outliers with NaN:")
print(f"  Traffic missing:     {bridge_clean['traffic'].isnull().sum()}")
print(f"  Temperature missing: {bridge_clean['temperature'].isnull().sum()}")

# (d) Justification
print("\n--- Justification ---")
print("Traffic values above ~800 vehicles/hour are physically implausible for a")
print("single bridge sensor — they likely represent sensor glitches (e.g. max-value")
print("overflow). Temperature readings of -40°C or +60°C are far outside the normal")
print("range for a bridge deck in a temperate climate.")
print("Replacing with NaN (rather than dropping rows) preserves the temporal structure")
print("of the time series, allowing proper interpolation in the next task.")
````

````python
# ✅ EXAMPLE SOLUTION — Task 5 (one possible approach)

# --- Approach 1: Monthly averages ---
monthly_traffic = traffic_filled.resample('ME').mean().round(0)
monthly_temp = temp_filled.resample('ME').mean().round(1)

print("--- Monthly average traffic ---")
print(monthly_traffic)
print(f"\nBusiest month:  {monthly_traffic.idxmax().strftime('%B')} ({monthly_traffic.max():.0f} vehicles/hour)")
print(f"Quietest month: {monthly_traffic.idxmin().strftime('%B')} ({monthly_traffic.min():.0f} vehicles/hour)")

# Plot monthly temperature
monthly_temp.plot(
    title='Monthly Average Bridge Deck Temperature — 2023',
    ylabel='Temperature (°C)',
    xlabel='Month',
    marker='o',
    figsize=(10, 3)
)
````

````python
# --- Approach 2: Weekday vs. weekend ---
day_names = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
by_weekday = traffic_filled.groupby(traffic_filled.index.dayofweek).mean()
by_weekday.index = day_names
print("\n--- Average traffic by day of week ---")
print(by_weekday.round(0))

by_weekday.plot(
    kind='bar',
    title='Average Hourly Traffic by Day of Week',
    ylabel='Vehicles / hour',
    figsize=(8, 3)
)
````

````python
# --- Approach 3: 30-day rolling average ---
daily_traffic = traffic_filled.resample('D').mean()
rolling_30 = daily_traffic.rolling(window=30).mean()

ax = daily_traffic.plot(
    alpha=0.4, label='Daily mean', color='steelblue',
    title='Annual Traffic Trend — 30-Day Rolling Average',
    ylabel='Vehicles / hour',
    figsize=(12, 4)
)
rolling_30.plot(ax=ax, linewidth=2, label='30-day rolling avg', color='red')
ax.legend()

print("\n--- Pattern ---")
print("Traffic is consistently lower on weekends (Sat/Sun) compared to weekdays,")
print("reflecting the commuter-driven nature of the bridge. The 30-day rolling average")
print("shows relatively stable traffic throughout the year with no strong seasonal trend.")
````
